### Traffic perturbation
##### 
Traffic perturbation of a traffic network is the temporal 
modification of the physical characteristics of the 
traffic network or the conditions to which the users are subjected
or both of them resulting in an overall deterioration of the 
traffic query indices such as travel times and more. [1]

Reference : 
[1] M. Nogal, A. O’Connor, B. Caulfield, and B. Martinez-Pastor, “Resilience of traffic networks: From perturbation to recovery via a dynamic restricted equilibrium model,” Reliab. Eng. Syst. Saf., vol. 156, pp. 84–96, 2016.
