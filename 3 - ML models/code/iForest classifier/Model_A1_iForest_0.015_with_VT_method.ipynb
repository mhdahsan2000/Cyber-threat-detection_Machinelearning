### iForest classifier

The primary aim of the iForest classifier and is to obtain the records 
in contrast to finding the normal observations. The algorithm would 
split the records by selecting a feature by choosing a value between
the maximum and the minimum values with respect to the chosen 
feature. 

Most of the existing model-based approaches to anomaly detection 
would construct a profile of the normal instances; it then identifies 
the instances that do not confirm the normal profile as anomalies. 

The use of isolation would enable the proposed model or iForest 
to exploit the sub sampling to an extent that is not feasible in the 
existing methods creating an algorithm which has a linear time 
complexity with a low constant and low memory requirement. 

Experimental Design : Tuning only one paramter of the Sklearn iForest 
model which is "contamination", the contamination parameter would show the proportion of observations that can be considered as outliers.

### Train

```python
# read in the train input dataset (After pre-processing and Feature selection.)
df_train = pd.read_csv('train_data_VT_A1.csv', sep=',')
```

```python

"""
The train and test dataset that were provided were pre-processed and 
feature selection was also done with two different methods 
namely Variance treshold (VT) and Principle Component Analysis (PCA)
before taking as input parameter to the machine learning classifier 
for classification inorder to create models.

The training and test data dataset that was read as input 
parameter to this code, have undergone the above two processes. 

1. Pre-processing or Data cleaning (to remove noise)
2. Feature selection using :
        i. Variance Treshold
        ii. Principal Component Analysis (PCA). 

"""
```

```python
df_train.head()
```

```python
# We set the isolation parameter equal to 0.015 initially 
isln_frst = IsolationForest(contamination=0.015)
```

```python
# Fit the training data to the model. 
isln_frst.fit(df_train)
```

### Test

```python
# read in the test input dataset (After pre-processing and Feature selection.)
df_test = pd.read_csv('test_data_VT_A1.csv', sep=',')
```

```python
predict_test = isln_frst.predict(df_test)

# Define the decision function
dscn_function = isln_frst.decision_function(df_test)


df_test['Model_Anomaly'] = predict_test
df_test['Model_Score'] = dscn_function
```

```python
# count the anomaly samples 
df_test['Model_Anomaly'].value_counts() # The value -1 in the results would correspond to the number of anomalies. 
```

```python
Anomalies_samples = df_test.loc[df_test['Model_Anomaly'] == -1]
```

```python
Anomalies_samples.head()
```

```python
test_original = pd.read_csv('testdata_cleaned.csv', sep=',')
```

```python
origin_feat_anomalies = test_original[test_original.index.isin(Anomalies_samples.index)]
```

```python
origin_feat_anomalies = origin_feat_anomalies.merge(Anomalies_samples[['Model_Score']], left_index = True, right_index=True)


```
